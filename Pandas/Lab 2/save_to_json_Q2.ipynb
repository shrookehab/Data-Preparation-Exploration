```python
import pandas as pd
```

```python
club_df = pd.read_csv('epldata_final.csv')
club_df.head()
```

```python
club_df.info()
```

```python
club_df = club_df.dropna(how='any')
club_df.head()
```

```python
club_df.info()
```

```python
club_df.groupby('club').mean()
```

```python
club_df.groupby('club')['age'].mean()
```

```python
club_df.groupby('club')['market_value'].mean()
```

```python
club_df.groupby('club')['age', 'market_value'].mean()
```
