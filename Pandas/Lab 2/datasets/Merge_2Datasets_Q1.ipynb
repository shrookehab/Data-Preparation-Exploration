```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
top_scores_df = pd.read_csv('Top_Scorers_dataset.csv')
top_scores_df.head()
```

```python
plyers_df = pd.read_json('players_stats.json')
plyers_df.head()
```

```python
merge_df = pd.merge(left=top_scores_df, right=plyers_df, left_on='Name', right_on='Name')
merge_df.head()
```

```python
merge_df = pd.merge(left=top_scores_df, right=plyers_df, on='Name')
merge_df.head()
```

```python
plt.scatter(merge_df['Appearances'], merge_df['Goals'])
plt.title('Relation between number Appearances and the number of scored goals')
plt.xlabel('Appearances')
plt.ylabel('Goals')
plt.show()
```
