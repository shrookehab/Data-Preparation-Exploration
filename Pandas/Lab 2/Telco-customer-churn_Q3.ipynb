```python
import pandas as pd
```

```python
customer_df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv', index_col=0)
customer_df.head()
```

```python
customer_df.info()
```

```python
customer_df.duplicated()
```

```python
customer_df[customer_df.duplicated()]
```

```python
customer_df = customer_df.drop_duplicates()
```

```python
customer_df.info()
```

```python
customer_df['TotalCharges'] = pd.to_numeric(customer_df['TotalCharges'], errors='coerce')#will leave the missing with NaN
customer_df.info()
```

```python
pd.get_dummies(customer_df['TotalCharges'])
```

```python
customer_df.info()
```

```python
filling_values = {"TotalCharges": customer_df['TotalCharges'].mean()}
customer_df = customer_df.fillna(value=filling_values)
customer_df
```

```python
customer_df.info()
```

```python
customer_df = pd.concat([customer_df,pd.get_dummies(customer_df[['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'PaymentMethod']])], axis=1)
customer_df.head()
```

```python
customer_df.info()
```

```python
del customer_df['gender']
del customer_df['Partner']
del customer_df['Dependents']
del customer_df['PhoneService']
del customer_df['MultipleLines']
del customer_df['InternetService']
del customer_df['OnlineSecurity']
del customer_df['OnlineBackup']
del customer_df['DeviceProtection']
del customer_df['TechSupport']
del customer_df['StreamingTV']
del customer_df['StreamingMovies']
del customer_df['PaperlessBilling']
del customer_df['PaymentMethod']
```

```python
customer_df = customer_df.drop(columns=['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'PaymentMethod'])
```

```python
customer_df.info()
```

```python
customer_df
```

```python
ordinal_encoding_mapping = {"Month-to-month":1, "One year":2, "Two year":3}
customer_df["Contract"] = customer_df["Contract"].replace(ordinal_encoding_mapping)
customer_df
```

```python
customer_df.info()
```

```python
customer_df['Contract'] = pd.to_numeric(customer_df['Contract'], errors='coerce')#will leave the missing with NaN
customer_df.info()
```

```python
ordinal_encoding_mapping = {"No":0, "Yes":1}
customer_df["Churn"] = customer_df["Churn"].replace(ordinal_encoding_mapping)
customer_df
```

```python
customer_df.info()
```

```python
customer_df['Churn'] = pd.to_numeric(customer_df['Churn'], errors='coerce')#will leave the missing with NaN
customer_df.info()
```

```python
churn_df = pd.DataFrame(data=customer_df["Churn"])
churn_df
```

```python
del customer_df["Churn"]
customer_df
```

```python
customer_df = (customer_df - customer_df.min()) / (customer_df.max() - customer_df.min())
customer_df
```

```python
customer_df.info()
```
